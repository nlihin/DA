## Selecting data

Select specific rows and/or columns using iloc (index based location):
(note that only the last row will show)

```python
# Rows:
data.iloc[0] # first row of data frame  
data.iloc[1] # second row of data frame  
data.iloc[-1] # last row of data frame  
# Columns:
data.iloc[:,0] # first column of data frame  
data.iloc[:,1] # second column of data frame  
data.iloc[:,-1] # last column of data frame  
```

The : operator 

 - when used alone it means "everything"

- also used to indicate a range of values


```python
#Rows and columns
data.iloc[0:5] # first five rows of dataframe
data.iloc[:, 0:2] # first two columns of data frame with all rows
data.iloc[[0,3,6,24], [0,5,6]] # 1st, 4th, 7th, 25th row + 1st 6th 7th columns.
```

## Conditional selection

Let's select task #1901:

```python
data.loc[:,'task_id'] == 1901
```

This creates a series of true/false 

We can insert this into data to select only that task:

```python
data[data.loc[:,'task_id'] == 1901]
```

Another way:

```python
data.loc[data.task_id == 1901]
```

Select two tasks:

```python
two_tasks = data.loc[(data.task_id == 1901) | (data.task_id == 1902)]
two_tasks
```
